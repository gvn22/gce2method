```python
def ranks(array):
    return len(np.where(array > 1e-9)[0])

def moderanks(array):
    nt = np.shape(array)[2]
    nm = np.shape(array)[1]
    
    R = np.reshape(np.zeros(nt*nm),(nt,nm))
    for t in np.arange(nt):
        for m in np.arange(nm):
            R[t,m] = ranks(array[:,m,t])

    return R
```
